```python
import numpy as np
import ta
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_data = yf.Ticker('TATAMOTORS.NS').history(period="1y")
stock_data.head()
```

```python
stock_data.to_csv('TATA_historical_data.csv')
df = pd.read_csv('TATA_historical_data.csv')
df.head()
```

```python
#using ta library to get rsi and ema
df["macd_data"] = ta.trend.macd(df['Close'])
df["rsi_data"] =  ta.momentum.rsi(df['Close'])
```

```python
# Preallocate the array for VMA
VMA = np.zeros(len(df))

# Loop to calculate the moving average
for i in range(len(df)):
    Volumes = df.loc[:i, "Volume"]  # Select the volume data up to the current row
    VMA[i] = np.mean(Volumes)  # Calculate the mean and assign it to the VMA array

# Add the VMA to the DataFrame as a new column
df["VMA"] = VMA
```

```python
df.head()
```

```python
df["macd_ema_9"] = ta.trend.macd_signal(df['Close'])
df.tail()
```

```python
df[df["macd_data"]<df["macd_ema_9"]].head()
```

```python
st_capital=100000
stocks=0
trade_call=np.array([])
value_at_eod=np.array([])
capital_eod=np.array([])
stock_holding=np.array([])
strong_trade=20
trade=10
for i in range(len(df)):
    if df.iloc[i,8]>0 and df.iloc[i,8]>df.iloc[i,11] and df.iloc[i,9]<70 and df.iloc[i,5]>df.iloc[i,10] and df.iloc[i,4]>df.iloc[i,1] and (st_capital>strong_trade*df.iloc[i,4]):
         stocks+=strong_trade
         st_capital-=strong_trade*df.iloc[i,4]
         trade_call=np.append(trade_call,1)  
    elif df.iloc[i,9]<40 and (st_capital>trade*df.iloc[i,4]):
        stocks+=trade
        st_capital-=trade*df.iloc[i,4]
        trade_call=np.append(trade_call,1)
    elif df.iloc[i,9]>70 and df.iloc[i,5]>df.iloc[i,10] and df.iloc[i,1]>df.iloc[i,4] and stocks>trade:
        stocks-=trade
        st_capital+=trade*df.iloc[i,4]
        trade_call=np.append(trade_call,-1)
    elif df.iloc[i,8]<0 and df.iloc[i,5]>df.iloc[i,10] and df.iloc[i,1]>df.iloc[i,4] and stocks>strong_trade:
        stocks-=strong_trade
        st_capital+=strong_trade*df.iloc[i,4]
        trade_call=np.append(trade_call,-1)
    else:
        trade_call=np.append(trade_call,0)
    value_at_eod=np.append(value_at_eod,st_capital+(stocks*df.iloc[i,4]))
    capital_eod=np.append(capital_eod,st_capital+(stocks*df.iloc[i,4]))
    stock_holding=np.append(stock_holding,stocks)
```
